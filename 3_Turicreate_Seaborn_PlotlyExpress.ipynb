**Question 1:** Write a function that returns the most popular book in each subject. Use it to find the most popular book in
    the _Mystery Fiction_ subject (15pt)

````python
# to make things run faster, we create smaller SFrame
sf2 = sf.copy()
sf2 = sf2.select_columns(['BibNum','year','Subjects','Title', 'Author', 'ItemCount'])
# remove unnecessary symbols
sf2['subject_list'] = sf2['Subjects'].apply(lambda s: s.split(","))
sf2['subject_list'] = sf2['subject_list'].apply(lambda l: [subject.strip() for subject in l])
# remove old subjects column
sf2 = sf2.remove_column('Subjects')
# convert list to different rows
sf2 = sf2.stack("subject_list", new_column_name="subject")
````

````python
import turicreate.aggregate as agg

def most_popular_book_per_subject(sf_to_use, subject_name):
    # This function retruns the most popular book in the given SFrame for a subject name
    sf_new_col = sf_to_use.copy()
    sf_new_col = sf_new_col[['Title', 'subject', 'ItemCount']]
    sf_new_col = sf_new_col[sf_new_col['subject'] == subject_name]
    sf_new_col.materialize()
    g = sf_new_col.groupby(['Title'],{'Sum': agg.SUM('ItemCount')})
    # For an SFrame that is lazily evaluated, force the persistence of the SFrame to disk, committing all lazy evaluated operations.
    g.materialize()
    g = g.sort("Sum", ascending=False)[0]
    top_book_title = g['Title']
    top_sum = g['Sum']
    return top_book_title, top_sum
    
subject_of_intrest = 'Mystery fiction'
book_title, book_count = most_popular_book_per_subject(sf_to_use=sf2, subject_name=subject_of_intrest)
print(f"The most popular book in the {subject_of_intrest} is: \"{book_title}\" (Count: {book_count})")
````

**Question 2**: Find the top-10 fiction subjects (5pt), and use Seaborn to visualize the number of items over time for each subject (15pt)
    

````python
# Find the top-10 fiction subjects
sf3 = sf2.copy()
# we need the year column for the visualization
sf3 = sf3.select_columns(['year', 'subject', 'ItemCount'])
sf3 = sf3[sf3['subject'].apply(lambda s: 'fiction' in s.lower().split(" "))]
g3 = sf3.groupby('subject', {'Max': agg.SUM('ItemCount')})
g3 = g3.sort("Max", ascending=False)
# g3 = g3[g3['subject'].apply(lambda s: 'fiction' in s.split(" "))]
top_subjects = g3['subject'][:10]
top_counts = g3['Max'][:10]
print('Top-10 fiction subjects:')
i = 1
for subject, subject_max in zip(top_subjects, top_counts):
  print(f'{i}. {subject} (Count: {subject_max})')
  i += 1
````

````python
# use Seaborn to visualize the number of items over time for each subject
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 

sns.set() # set style to seaborn defaults
i = 1
for subject, subject_max in zip(top_subjects, top_counts):
    # Initialize the matplotlib figure
    f, ax = plt.subplots(figsize=(15, 10))
    plt.xticks(rotation=90)
    curr_sub_sf = sf3[sf3['subject'] == subject]
    g = curr_sub_sf.groupby(['subject', 'year'], {'Number of items': agg.SUM('ItemCount')}).sort("year").to_dataframe()
    sns.barplot(x="year", y="Number of items", data=g).set_title(f"Subject #{i}: {subject} (Count: {subject_max})")
    sns.despine(left=True, bottom=True)
    i += 1
plt.tight_layout() # automatically adjusts subplot params so that the subplot(s) fits in to the figure area 
````


**Question 3:** Write a function that gets as input a subject name. The function returns the top-20 most _popular_ author for the input subject (15pt).



````python
# Find top-20 most popular author for the input subject

def top_k_authors_per_subject(sf_to_use, subject_name, k):
    sf_new_col = sf_to_use[['subject', 'ItemCount', 'Author']]
    sf_new_col = sf_new_col[sf_new_col['subject'] == subject_name]
    g = sf_new_col.groupby('Author', {'Sum': agg.SUM('ItemCount')})
    g = g[g['Author']!='']
    g = g.sort('Sum', ascending=False)[:k]
    return g['Author'], g['Sum']

subject_of_intrest = 'Mystery fiction'
top_authors_per_subject, top_authors_count = top_k_authors_per_subject(sf_to_use=sf2, subject_name='Mystery fiction', k=20)
print(f"The most popular authors for {subject_of_intrest} are:")
i = 1
for subject, subject_max in zip(top_authors_per_subject, top_authors_count):
  print(f'{i}. {subject} (Count: {subject_max})')
  i += 1
````

````python
# We will start by converting the XML files into a JSON file:
!mkdir ./datasets
!mkdir ./datasets/BIU-Blog-Authorship
!wget -O ./datasets/BIU-Blog-Authorship/blogs.zip http://www.cs.biu.ac.il/~koppel/blogs/blogs.zip
````

````python
#first we create a directory to put the JSON files
import os
import json
!unzip /content/datasets/BIU-Blog-Authorship/*.zip  -d /content/datasets/BIU-Blog-Authorship/
blogger_xml_dir = "/content/datasets/BIU-Blog-Authorship/blogs"

#We create a short code which parse the XML and convert it to JSON files
def get_posts_from_file(file_name):
    posts_dict = {}
    txt = open(file_name, "r",  encoding="utf8", errors='ignore').read()
    txt = txt.replace("&nbsp;", " ")
    for p in txt.split("</post>"):
        if "<post>" not in p or "<date>" not in p:
            continue
        post = p.split("<post>")[1].strip()
        dt = p.split("</date>")[0].split("<date>")[1].strip()
        posts_dict[dt] = post

    return posts_dict
            

def blogger_xml_to_json(file_name):
    l = file_name.split("/")[-1].split(".")
    if len(l) != 6:
        raise Exception("Could not analyze file f{file_name} - Length %s" % len(l) )
    j = {"id": l[0], "gender": l[1], "age":int(l[2]), "topic":l[3], "sign": l[4], "posts": get_posts_from_file(file_name)}
    return j

# converting all the XMLs to a single large JSON file
all_jsons = []
for p in os.listdir(blogger_xml_dir):
    if not p.endswith(".xml"):
        continue
    j = blogger_xml_to_json(f"{blogger_xml_dir}/" + p)
    all_jsons.append(j)
json.dump(all_jsons, open(f"{blogger_xml_dir}/all_bloggers.json","w" ))
````

Now let's load the JSON file to an SFrame object using the read_json function:

````python
import turicreate as tc
import turicreate.aggregate as agg


blog_sf = tc.SFrame.read_json(f"{blogger_xml_dir}/all_bloggers.json")
blog_sf
````

## Calculate the average and median age of bloggers in each topic by gender(5pt)

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def avg_and_median(sf_to_use, main_column, agg_column):
  g1 = sf_to_use.groupby(main_column, {'average_age': agg.MEAN(agg_column)})
  g2 = sf_to_use.groupby(main_column, {'median_age': agg.QUANTILE(agg_column, 0.5)})
  g2['median_age'] = g2['median_age'].apply(lambda x: x[0])
  g1['median_age'] = g2['median_age']
  return g1

male_sf = blog_sf[blog_sf['gender'] == 'male']
male_sf_results = avg_and_median(sf_to_use=male_sf, main_column='topic', agg_column='age')
female_sf = blog_sf[blog_sf['gender'] == 'female']
female_sf_results = avg_and_median(sf_to_use=female_sf, main_column='topic', agg_column='age')
````

````python
male_sf_results['gender'] = 'male'
female_sf_results['gender'] = 'female'
# print(male_sf_results)
# print(female_sf_results)
two_gender = male_sf_results.append(female_sf_results)
print(two_gender)
````

````python
sign_sf_results = avg_and_median(sf_to_use=blog_sf , main_column='sign', agg_column='age')
print(sign_sf_results)
````

## Visualize the age distributions of bloggers according to their topic and gender using Seaborn and PlotlyExpress(15pt)
**I used the violin graph to visualize the age distributions of bloggers according to their topic and gender:**
Violin plots are similar to box plots, except that they also show the probability density of the data at different values, usually smoothed by a kernel density estimator. Typically a violin plot will include all the data that is in a box plot: a marker for the median of the data; a box or marker indicating the interquartile range; and possibly all sample points, if the number of samples is not too high (Wikipedia [Violin Plot](https://en.wikipedia.org/wiki/Violin_plot)).


````python
# list of intresting topics
num_of_topics = 7
top_topics = blog_sf.groupby('topic', {'Count': agg.COUNT()}).sort("Count", ascending=False)[:num_of_topics].to_dataframe()['topic'].tolist()
print(top_topics)

final_df = blog_sf.copy()
final_df['age'] = final_df['age'].astype(float)
final_df['gender'] = final_df['gender'].astype(str)
final_df['topic'] = final_df['topic'].astype(str)
final_df = final_df[['age', 'gender', 'topic']]
final_df = final_df.to_dataframe()
final_df = final_df[final_df['topic'].isin(top_topics)]
````

````python
import turicreate as tc
import matplotlib.pyplot as plt
import seaborn as sns
import turicreate.aggregate as agg
%matplotlib inline

SJR_PATH = "./datasets/sjr/data/sjr2018.csv"
sjr_sf = tc.SFrame.read_csv(SJR_PATH, delimiter=";")
````

## Using the SJR Journal Ranking of 2018 dataset and Seaborn, visualize the H-index distributions according to the SJR Best Quartile (15pt)

````python
quartile = sjr_sf['SJR Best Quartile'].unique()
for q in quartile:
    f, ax = plt.subplots(figsize=(10, 10)) # Initialize the matplotlib figure
    plt.xticks(rotation=90)
    curr_q_sf = sjr_sf[sjr_sf['SJR Best Quartile'] == q]
    g = curr_q_sf.groupby(['H index'], {'Count': agg.COUNT()}).to_dataframe()
    sns.distplot(g['H index']).set_title(f"SJR Best Quartile {q}:")
    sns.despine(left=True, bottom=True)
````

````python
import numpy as np
print(sjr_sf.column_names())
sjr_sf_float = sjr_sf.copy()
sjr_sf_float["Cites / Doc. (2years)"] = sjr_sf["Cites / Doc. (2years)"].apply(lambda s: int(float(s.replace(",", "."))+1))
sjr_sf_float.materialize()
# Select publishers
publisher_num = 10
top_publisher = sjr_sf_float.groupby('Publisher', {'Count': agg.SUM('Cites / Doc. (2years)')}).sort("Count", ascending=False)[:publisher_num]
print(top_publisher)
top_publisher = top_publisher.to_dataframe()['Publisher'].tolist()
top_publisher.remove('')
# Filtering
df_to_plot = sjr_sf_float[sjr_sf_float['Publisher'].apply(lambda x: x in top_publisher)].to_dataframe()
print('Cites / Doc. (2years) histograms values of each publisher:')
# Plot
sns.set(style="darkgrid")
g = sns.FacetGrid(df_to_plot, col="Publisher", sharex=False, sharey=False)
g.map(plt.hist, 'Cites / Doc. (2years)', bins=range(0, 10), color="steelblue")
plt.tight_layout()
plt.show()
````
